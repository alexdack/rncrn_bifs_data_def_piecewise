```python

def curv_pos(t):
    return np.transpose(np.array([2+ (np.cos(t)/(np.sin(t)**2 +1)), 2+((np.cos(t)*np.sin(t))/(np.sin(t)**2 +1))]))
```

```python
def create_attractor_two_points_second(first_point, second_point, d, k1):
    v = second_point-first_point
    v_norm = v / np.linalg.norm(v)
    norm_vec = np.matmul(np.array([[0, -1], [1,0]]), np.transpose(v) )
    unit_norm_vec = norm_vec / np.linalg.norm(norm_vec)
    Q_plus = first_point + d*unit_norm_vec
    Q_neg = first_point - d*unit_norm_vec
    inputs = np.array([first_point, Q_plus, Q_neg]);
    k2 = k1*np.exp(d)
    targets = np.array([k1*v_norm,  k1*v_norm - k2*unit_norm_vec , k1*v_norm + k2*unit_norm_vec ]);
    return (inputs, targets)

def create_attractor_two_points_padd(first_point, second_point, d, k1, num_padd):
    v = second_point-first_point
    v_norm = v / np.linalg.norm(v)
    norm_vec = np.matmul(np.array([[0, -1], [1,0]]), np.transpose(v) )
    unit_norm_vec = norm_vec / np.linalg.norm(norm_vec)
    
    inputs = np.array([first_point]);
    targets = np.array([k1*v_norm ]);
          
    for i in range(1, num_padd+1):
        Q_plus = first_point + i*d*unit_norm_vec
        Q_neg = first_point - i*d*unit_norm_vec
        inputs = np.concatenate((inputs, np.array([Q_plus, Q_neg])))
        k2 = k1*np.exp(i*d)
        targets = np.concatenate((targets, np.array([k1*v_norm - k2*unit_norm_vec , k1*v_norm + k2*unit_norm_vec])))

    return (inputs, targets)
    
d = 0.01
k1 = 1
num_padd = 5
data_point_per = 1 + 2*(num_padd)

first_point = np.array([1, 1])
second_point = np.array([2, 2])
third_point = np.array([2, 3])

t_arr = np.arange(0, 2*np.pi, 0.01)
points_arr = curv_pos(t_arr)

N = points_arr.shape[0]
res_input = np.zeros((data_point_per*N,2));
res_targets = np.zeros((data_point_per*N,2));


for i in np.arange(0, N-1, 1):
    inputs, targets = create_attractor_two_points_padd(points_arr[i], points_arr[i+1], d, k1, num_padd)
    #print(inputs.shape)
    res_input[i*data_point_per:(i+1)*data_point_per, :] = inputs
    res_targets[i*data_point_per:(i+1)*data_point_per, :] = targets
    
inputs, targets = create_attractor_two_points_padd(points_arr[N-1], points_arr[0], d, k1, num_padd)
res_input[(N-1)*data_point_per:N*data_point_per, :] = inputs
res_targets[(N-1)*data_point_per:N*data_point_per, :] = targets    
    


plt.quiver(res_input[:,0], res_input[:,1], res_targets[:,0], res_targets[:,1], color='green')


inputs_a = np.transpose(res_input)
targets_a = np.transpose(res_targets)
```
