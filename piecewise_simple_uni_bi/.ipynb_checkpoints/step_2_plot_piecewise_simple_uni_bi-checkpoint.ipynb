```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
import matplotlib as mpl

# imports custom RNCRN functions that are used throughout multiple  notebooks 
import os
import sys
from pathlib import Path

module_path = str(Path(os.path.abspath('')).parent)
if module_path not in sys.path:
    sys.path.append(module_path)
    
from functions.target_systems import *
from functions.RNCRN_tools import * 
from functions.RNCRN_phase_plane import *
from functions.create_attractors_tools import * 
from functions.create_chimera_tools import initialize_static_execs_data, initialize_single_RNCRN_static_exec, merge_static_execs_multi_state_data
from functions.RNCRN_phase_line import create_training_data_1D_from_ODE, compute_quasi_static_line_over_state_space_static_execs
from functions.RNCRN_train import * 

save_data = True
```

```python
# sets the parameters of the target dynamics 
target_dynamics_a = lambda x: uni_stable(x, a=5, b=6)
target_dynamics_b = lambda x: cubic_bi_stable(x, r1=2, r2=5, r3=8)
```

```python
# sets intervals of executive species compact subsets (i.e. K_1, K_2)
lower_limit = 1
upper_limit = 9
step_size = 0.01

# creates training data from ODE
inputs_a, targets_a = create_training_data_1D_from_ODE(target_dynamics_a, lower_limit, upper_limit, step_size )
inputs_b, targets_b = create_training_data_1D_from_ODE(target_dynamics_b, lower_limit, upper_limit, step_size )

# defines the number of parameters
number_of_static_exec_species = 1

# defines map between parameter species value and target dynamics
static_exec_state_a = [1]
static_exec_state_b = [0]

# initialize training data for the parameter species 
exec_inputs_a = initialize_static_execs_data(number_of_static_exec_species, inputs_a, static_exec_state_a)
exec_inputs_b = initialize_static_execs_data(number_of_static_exec_species, inputs_b, static_exec_state_b)
```

```python
# name of model to plot
model_name= r"models/example_piecewise_simple_uni_bi"

# call a function to open the saved model and unpack the parameters to their named matrices 
params, number_of_exec_species, number_chemical_perceptrons, number_of_static_exec_species = unpack_mat_model_static_exec(model_name)
```

```python
# calculates the quasi-static vector field for the RNCRN given the parameter matrices  
pred_a = compute_quasi_static_line_over_state_space_static_execs(inputs_a, exec_inputs_a, params[0], params[1], params[2], jnp.abs(params[3]), jnp.abs(params[4]), jnp.abs(params[5]), params[6])

# calculates the quasi-static vector field for the RNCRN given the parameter matrices  
pred_b = compute_quasi_static_line_over_state_space_static_execs(inputs_b, exec_inputs_b, params[0], params[1], params[2], jnp.abs(params[3]), jnp.abs(params[4]), jnp.abs(params[5]), params[6])
```

```python
if save_data:
    np.savetxt("data/target_dynamics_a.csv", np.asarray([ inputs_a[0,:], targets_a[0,:]]), delimiter=",")
    np.savetxt("data/target_dynamics_b.csv", np.asarray([ inputs_b[0,:], targets_b[0,:]]), delimiter=",")
    np.savetxt("data/rncrn_dynamics_a.csv", np.asarray([ inputs_a[0,:], pred_a[0,:]]), delimiter=",")
    np.savetxt("data/rncrn_dynamics_b.csv", np.asarray([ inputs_b[0,:], pred_b[0,:]]), delimiter=",")
```

```python
# reformat params to be work with CRN saving framework 
alpha_mat, omega_mat, bias_vec, beta_vec, gamma_vec, tau_vec, static_omega_mat = params 
new_omega_mat = np.concatenate([omega_mat, static_omega_mat], axis=1)
new_params = (alpha_mat, new_omega_mat, bias_vec, beta_vec, gamma_vec, tau_vec)

# sets a time-scale (\mu parameter) for the chemical reaction network 
time_scale = 0.1

# sets all initial conditions to zero (overwritten later)
inits = np.zeros(number_of_exec_species+number_chemical_perceptrons)

# sets all parameter species initial conditions to zero (overwritten later)
number_of_control_params = 1
control_inits = np.zeros(number_of_control_params)

# sets file location to store the compiled chemical reaction network
crn_file = 'CRNs/piecewise_simple_uni_bi.txt'

# compiles the chemical reaction network and saves as .txt
save_crn(crn_file,  convert_static_exec_RNCRN_params_to_CRN(new_params, time_scale, inits, control_inits, number_of_exec_species, number_chemical_perceptrons, number_of_control_params) )

# loads the chemical reaction network file from .txt and extracts the kinetic and stoichiometric data
species, reaction_rates, react_stoch, prod_stoch, stoch_mat, number_species, number_reactions, initial_concs = read_crn_txt(crn_file)

# creates a tuple of kinetic and stoichiometric data for the trajectory simulation 
args_tup_rncrn = (reaction_rates, react_stoch, stoch_mat,)
```

```python
# numerical integration of the RNCRN 
# the time length of the trajectory
tFinal = 1

# handler function for setting new initial conditions in chemical reaction networks 
inits_rncrn_a_2, id_dict_a = new_initial_conditions(initial_concs, species, {'X_1':2, 'L_1': 1})

# simulates the time-trajectory of the chemical reaction netwrok 
sol_rncrn_a_2 = solve_ivp(stoch_mat_to_mass_action, [0, tFinal], inits_rncrn_a_2, args=args_tup_rncrn, rtol=10e-8)

# handler function for setting new initial conditions in chemical reaction networks 
inits_rncrn_a_3, id_dict_a = new_initial_conditions(initial_concs, species, {'X_1':3, 'L_1': 1})

# simulates the time-trajectory of the chemical reaction netwrok 
sol_rncrn_a_3 = solve_ivp(stoch_mat_to_mass_action, [0, tFinal], inits_rncrn_a_3, args=args_tup_rncrn, rtol=10e-8)

# handler function for setting new initial conditions in chemical reaction networks 
inits_rncrn_a_6, id_dict_a = new_initial_conditions(initial_concs, species, {'X_1':6, 'L_1': 1})

# simulates the time-trajectory of the chemical reaction netwrok 
sol_rncrn_a_6 = solve_ivp(stoch_mat_to_mass_action, [0, tFinal], inits_rncrn_a_6, args=args_tup_rncrn, rtol=10e-8)

# handler function for setting new initial conditions in chemical reaction networks 
inits_rncrn_a_7, id_dict_a = new_initial_conditions(initial_concs, species, {'X_1':7, 'L_1': 1})

# simulates the time-trajectory of the chemical reaction netwrok 
sol_rncrn_a_7 = solve_ivp(stoch_mat_to_mass_action, [0, tFinal], inits_rncrn_a_7, args=args_tup_rncrn, rtol=10e-8)
```
